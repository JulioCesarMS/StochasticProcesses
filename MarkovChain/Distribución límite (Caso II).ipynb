Supongamos que tenemos la siguiente matriz de transición

$$\textbf{P} =
\begin{pmatrix}
\frac{1}{2}  & \frac{1}{2}  &  0            &  0\\
       0     & \frac{1}{2}  &  \frac{1}{2}  &  0 \\
       0     &        0     & \frac{1}{2}   &  \frac{1}{2} \\
       0     &        0     &       0       & \frac{1}{2} 
\end{pmatrix}$$

Obtener $\lim_{n \rightarrow \infty} P_{i,j}^n$.

```r
plotmat(tm.tr,pos=c(1,2),name=c("AA","Aa","aa"),
arr.length=0.3,arr.width=0.1,box.col="lightblue",
box.lwd=1,box.prop=0.5,box.size=0.12,box.type="circle",
cex.txt=0.8,lwd=1,self.cex=0.6,self.shiftx=0.17,
self.shifty=-0.01)
```

```r
# Función para calcular la potencia de una matriz
mat.pow <- function(m, n){
  # inputs:
  #   m : matriz 
  #   n : potencia
  # outputs:
  #   Mn : matriz potencia n
  for(i in 1:n){
    if(i == 1){
      Mn <- m
    }else{
      Mn <- Mn%*%m
    }
  }
  return(Mn)
}
```

Probamos para $n$ de $1$ hasta $50$.

```r
# Declaranmos nuestra matriz
P = matrix(c(1/2, 1/2, 0, 0, 
             0, 1/2, 1/2, 0,
             0, 0 , 1/2, 1/2,
             0, 0, 0, 1/2), ncol=4, byrow = TRUE)
print(P)
```

```r
# Diagrama de transiciones
options(repr.plot.width=10, repr.plot.height=8)
plotmat(t(P),name=c("S0","S1","S2", 'S3'),
arr.length=0.3,arr.width=0.1,box.col="lightblue",box.lwd=1,box.prop=0.5,box.size=0.09,
box.type="circle",cex.txt=0.8,lwd=1,self.cex=0.3,
self.arrpos=0.3,self.shiftx=0.09,self.shifty=-0.05)
```

```r
n.max = 50
for(i in 1:n.max){
    Pn <- mat.pow(P, n=i)
    cat(paste('Matriz de probabilidades de transición a n= ', i, ' pasos \n', sep=''))
    print(Pn)
}
```
