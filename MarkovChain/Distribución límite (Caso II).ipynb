# Distribución límite (Caso II)

Supongamos que tenemos la siguiente matriz de transición

$$\textbf{P} =
\begin{pmatrix}
\frac{1}{2}  & \frac{1}{2}  &  0            &  0\\
       0     & \frac{1}{2}  &  \frac{1}{2}  &  0 \\
       0     &        0     & \frac{1}{2}   &  \frac{1}{2} \\
       0     &        0     &       0       & \frac{1}{2} 
\end{pmatrix}$$

Obtener $\lim_{n \rightarrow \infty} P_{i,j}^n$.

```r
# Función para calcular la potencia de una matriz
mat.pow <- function(m, n){
  # inputs:
  #   m : matriz 
  #   n : potencia
  # outputs:
  #   Mn : matriz potencia n
  for(i in 1:n){
    if(i == 1){
      Mn <- m
    }else{
      Mn <- Mn%*%m
    }
  }
  return(Mn)
}
```

Probamos para $n$ de $1$ hasta $50$.

```r
# Declaranmos nuestra matriz
P = matrix(c(1/2, 1/2, 0, 0, 
             0, 1/2, 1/2, 0,
             0, 0 , 1/2, 1/2,
             0, 0, 0, 1/2), ncol=4, byrow = TRUE)
print(P)
```

```r
n.max = 50
for(i in 1:n.max){
    Pn <- mat.pow(P, n=i)
    cat(paste('Matriz de probabilidades de transición a n= ', i, ' pasos \n', sep=''))
    print(Pn)
}
```
